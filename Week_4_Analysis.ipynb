```python
import pandas as pd

# Load the dataset
df = pd.read_csv(r"C:\Users\gouth\OneDrive\Desktop\Student Performance Analysis\Data\student-mat.csv", sep=";")

# Display the first 5 rows
df.head()
```

```python
# Analyze the impact of study time on final grades
studytime_analysis = df.groupby("studytime")["G3"].mean()

print("Average Final Grade by Study Time")
print(studytime_analysis)
```

```python
# Analyze the impact of mother's education on final grades
mother_analysis = df.groupby("Medu")["G3"].mean()

print("Average Final Grade by Mother's Education")
print(mother_analysis)
```

```python
# Analyze the impact of father's education on final grades
father_analysis = df.groupby("Fedu")["G3"].mean()

print("Average Final Grade by Father's Education")
print(father_analysis)
```

```python
import matplotlib.pyplot as plt

studytime_analysis.plot(kind="bar")

plt.title("Average Final Grade by Study Time")
plt.xlabel("Study Time")
plt.ylabel("Average Final Grade (G3)")
plt.show()
```

```python
comparison = pd.DataFrame({
    "Mother": mother_analysis,
    "Father": father_analysis
})

comparison.plot(kind="bar")

plt.title("Average Final Grade by Parents' Education")
plt.xlabel("Education Level")
plt.ylabel("Average Final Grade (G3)")
plt.show()
```
