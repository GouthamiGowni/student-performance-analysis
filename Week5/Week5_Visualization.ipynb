```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv(
    r"C:\Users\gouth\OneDrive\Desktop\Student Performance Analysis\Data\student-mat.csv",
    sep=";"
)

# Display first five rows
df.head()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["G3"], bins=10, kde=True)

plt.title("Distribution of Final Grades (G3)")
plt.xlabel("Final Grade")
plt.ylabel("Number of Students")

plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.boxplot(y=df["G3"])

plt.title("Boxplot of Final Grades (G3)")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["G1"], bins=10, kde=True)

plt.title("Distribution of First Period Grades (G1)")
plt.xlabel("First Period Grade (G1)")
plt.ylabel("Number of Students")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["G2"], bins=10, kde=True)

plt.title("Distribution of Second Period Grades (G2)")
plt.xlabel("Second Period Grade (G2)")
plt.ylabel("Number of Students")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(data=df[["G1", "G2", "G3"]])

plt.title("Comparison of Grade Distributions")
plt.xlabel("Grade")
plt.ylabel("Score")

plt.show()
```
